# Parallel

```julia
# Julia has been designed for straightforward parallelization. 
# The first step is to add workers to a task
using Distributed
addprocs(2)

# Next, we can have a parallel loop
using SharedArrays
using Distributed
simulation = SharedArray{Float64}(100)
@distributed for i in 1:100
        simulation[i] = i
end
```

# Try Catch-Errors

```julia
function square(x)
        try
                sqrt(x)
        catch e
        println(e)
end 
end

square(2) # returns 1.41...
square("economics") # returns MethodError(sqrt, ("economics",), 0....)
```

# Plots

```julia
# A simple plot can be done with the Plots package
import Pkg; Pkg.add("Plots")
using Plots
x = 1:10
y = x.^2
plot(x,y)
```

```julia
square(x) = x^2
plot(square,1:10,title="A nice plot", label = "Square function",line = (:blue,0.9,3, :dot), xlabel = "x-axis", ylabel ="y-axis")
cube(x) = x^3
plot!(cube,1:10, label = "Cube function",line =(:red,0.9,3, :dot))
#savefig("figure1.pdf")
```

```julia
x = 1:10
y = x.^2
Plots.scatter(x,y,label = "3d plot",line =(:red,0.9,3, :dot))
```
