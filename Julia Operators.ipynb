## Logical Operators

Julia has all the most commonly udes logical operators:
- !  not && # and || # or
- ==  is equal?
- !==  is not equal?
- ===  is equal? (enforcing type 2===2.0 is false
- '>'  bigger than
- '>='  bigger or equal than < # less than
- <=  less or equal than

```julia
# Logical Operators can be combined as follows
3 > 2 && 4<=8 || 7 < 7.1
```

```julia
#Note: Logical Operators in Julia are lazy --> they are only evaluated when needed
2 > 3 && println("I am lazy")
#The evaluation does not proceed over 2 > 3 as the statment is false
```

## Boolean Operators and Ascertain Funcitons

Julia includes all the boolean operators:
- ~    bitwise not
- &    bitwise and
- |    bitwise or
- xor    bitwise xor (also typed by \xor or \veebar + tab) >> # right bit shift operator
- <<    left bit shift operator
- '>>>'  unsigned right bit shift operator

And almost all the ascertain function
- isa(a,Float64)
- isnumeric(a)
- iseven(a)
- isodd(a)
- ispow2(a)
- isfinite(a)
- isinf(a)
- isnan(a)

Note: all of the aboce ascertain function have their converse, starting with !
